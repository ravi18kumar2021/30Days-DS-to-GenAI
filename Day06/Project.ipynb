```python
# first replace '?' with 0 to make int datatype, then fill 0 values with the mean or median of the horsepower column
df[df['horsepower'] == '?']
```

```python
df.loc[df[df['horsepower'] == '?'].index, 'horsepower'] = 0
df['horsepower'] = df['horsepower'].astype(int)
df['horsepower'].describe()
```

```python
# let's fill 0 with median value
df.loc[df[df['horsepower'] == 0].index, 'horsepower'] = df['horsepower'].median()
```

```python
df['horsepower'].describe()
```

```python
df.columns
```

```python
# correlation with target variable (miles per gallon)
df.corr(numeric_only=True)['mpg'].sort_values(ascending=False)
```

### Key Points

- Cylinders are negative correlated with mpg. It indicates that the larger engines are less fuel-efficient.
- Weights are also negative correlated with mpg. It indicates that heavy cars generally consume more fuel.
