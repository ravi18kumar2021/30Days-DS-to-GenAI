```python
# Before imputing values, first convert data types of particular columns
df['Glucose'] = df['Glucose'].astype(float)
df['BloodPressure'] = df['BloodPressure'].astype(float)
df['Insulin'] = df['Insulin'].astype(float)
```

```python
# impute 0 values with mean of columns plasma glucose, blood pressure, bmi and insulin
df.loc[df[df['Glucose'] == 0].index, 'Glucose'] = df['Glucose'].mean()
df.loc[df[df['BloodPressure'] == 0].index, 'BloodPressure'] = df['BloodPressure'].mean()
df.loc[df[df['Insulin'] == 0].index, 'Insulin'] = df['Insulin'].mean()
df.loc[df[df['BMI'] == 0].index, 'BMI'] = df['BMI'].mean()
```
