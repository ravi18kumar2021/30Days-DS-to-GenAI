### Grouping and Aggregation

```python
df = pd.DataFrame({
    'Product': ['Laptop', 'Pizza', 'Milk', 'Shirt', 'Refrigerator', 'Jeans', 'Egg', 'Burger'],
    'Category': ['Electronics', 'Food', 'Food', 'Clothes', 'Electronics', 'Clothes', 'Food', 'Food'],
    'Amount': [35000, 300, 50, 500, 20000, 1200, 40, 100]
})
df
```

```python
df.groupby('Category').count()
```

```python
df.groupby('Category')['Amount'].mean()
```

```python
df.groupby('Category')['Amount'].idxmax()
```

```python
df.loc[df.groupby('Category')['Amount'].idxmax()]
```

### Merging and Joining

```python
products_df = pd.DataFrame({
    'ProductID': [101, 102, 103, 104, 105, 106],
    'Product': ['Laptop', 'Pizza', 'Milk', 'Shirt', 'Refrigerator', 'Jeans'],
    'Category': ['Electronics', 'Food', 'Food', 'Clothes', 'Electronics', 'Clothes']
})

sales_df = pd.DataFrame({
    'SaleID': [1, 2, 3, 4, 5, 6],
    'ProductID': [101, 102, 104, 106, 107, 103],
    'UnitsSold': [5, 20, 7, 3, 2, 10],
    'TotalSale': [175000, 6000, 3500, 3600, 2000, 500]
})
```

```python
products_df
```

```python
sales_df
```

```python
# inner join
pd.merge(left=products_df, right=sales_df, on='ProductID', how='inner')
```

```python
# left join
pd.merge(left=products_df, right=sales_df, on='ProductID', how='left')
```

```python
# right join
pd.merge(left=products_df, right=sales_df, on='ProductID', how='right')
```

```python
# outer join
pd.merge(left=products_df, right=sales_df, on='ProductID', how='outer')
```

### Missing Values

```python
df = pd.DataFrame({
    'Product': ['Laptop', 'Pizza', 'Milk', 'Shirt', 'Refrigerator', 'Jeans', 'Burger'],
    'Category': ['Electronics', 'Food', 'Food', np.nan, 'Electronics', 'Clothes', 'Food'],
    'UnitsSold': [5, 20, np.nan, 7, 3, np.nan, 15],
    'Price': [70000, 300, 50, 500, 20000, 1200, np.nan]
})
```

```python
df
```

```python
# check missing values
df.isnull().sum()
```

```python
df[df['Category'].isnull()]
```

```python
df.loc[:, 'Category'] = df['Category'].fillna('Clothes')
df
```

```python
df[df['UnitsSold'].isnull()]
```

```python
df['UnitsSold'].mean()
```

```python
df.loc[:, 'UnitsSold'] = df['UnitsSold'].fillna(df['UnitsSold'].mean())
df
```

```python
df = df.dropna()
df
```
